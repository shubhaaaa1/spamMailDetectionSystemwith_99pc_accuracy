```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
data=spam['Category']
data1= encoder.fit_transform(data)
```

```python
X= spam['Message']
Y= data1
```

```python
print(X)
print(Y)
```

```python
#seprating test and train data
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
feature_ext = TfidfVectorizer(min_df=1,stop_words ='english',lowercase=True)
extracted_X_train= feature_ext.fit_transform(X_train)
extracted_X_test= feature_ext.transform(X_test)
```

```python
#training model with the trained data
model= LogisticRegression(class_weight='balanced')
model.fit(extracted_X_train,Y_train)
```

```python
#Checking the Accuracy of our Data
model_predict_train= model.predict(extracted_X_train)
trained_accuracy = accuracy_score(Y_train,model_predict_train)
model_predict_test=model.predict(extracted_X_test)
test_accuracy = accuracy_score(Y_test,model_predict_test)
```

```python
print('The Accuracy Score of the Trained data is : ',trained_accuracy)
print('The Accuracy Score of the Test Data is : ',test_accuracy)
```

```python
#Predictive System of the Project
input_data ="WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."
data3= feature_ext.transform([input_data])
prediction=model.predict(data3)
print(prediction)
if(prediction[0]==0):
    print('The Entered Email is a Ham Mail')
else:
    print('The Entered Email is a Spam Mail')
```
